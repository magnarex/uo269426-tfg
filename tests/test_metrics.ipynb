```python
model = Model().load(filename='test')
components = model.components
bins = model.src.bins
```

```python
xmin = bins.min()
xmax = bins.max()
size = (xmax-xmin)/(bins.size-1)
bin_mid = (bins[1:]+bins[:-1])/2

fig, ax = plt.subplots(1,1,figsize=(9,6))
ax.step(bin_mid,components.T,where='mid')
ax.legend([f'Comp. {i}' for i in range(components.shape[0])],fontsize=14)
```
